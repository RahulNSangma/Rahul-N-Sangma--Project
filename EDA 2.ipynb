```python
# Replace '?' with NaN
dataset.replace('?', pd.NA, inplace=True)
```

```python
# Impute or drop missing values
dataset.dropna(inplace=True)  # or use imputation techniques
```

## •Standard Scaling to numerical featues

```python
# Identify numerical features
numerical_features = dataset.select_dtypes(include=['int64', 'float64']).columns
```

## •Standard Scaling

```python
# Standard Scaling
standard_scaler = StandardScaler()
dataset[numerical_features] = standard_scaler.fit_transform(dataset[numerical_features])
```

```python
dataset[numerical_features]
```

## •Min-Max Scaling

```python
# Min-Max Scaling
min_max_scaler = MinMaxScaler()
dataset[numerical_features] = min_max_scaler.fit_transform(dataset[numerical_features])
```

```python
dataset[numerical_features]
```

# 2. Encoding Techniques:

```python
# Identify categorical features
categorical_features = dataset.select_dtypes(include=['object']).columns
```

```python
# One-Hot Encoding for categorical variables with < 5 categories
for feature in categorical_features:
    if dataset[feature].nunique() < 5:
        dataset = pd.get_dummies(dataset, columns=[feature], drop_first=True)
```

```python
# Update categorical_features after One-Hot Encoding
categorical_features = dataset.select_dtypes(include=['object']).columns # This line was added
# Label Encoding for categorical variables with > 5 categories
label_encoder = LabelEncoder()
for feature in categorical_features:
    if dataset[feature].nunique() >= 5:
        dataset[feature] = label_encoder.fit_transform(dataset[feature])
```

# 3. Feature Engineering:

```python
# Example features
dataset['age_income_ratio'] = dataset['age'] / dataset['fnlwgt']  # Ratio of age to final weight
dataset['education_num_squared'] = dataset['education_num'] ** 2  # Square education to capture non-linearity
```

```python
# Log transformation on skewed feature (fnlwgt)
dataset['log_fnlwgt'] = np.log(dataset['fnlwgt'] + 1)  # Adding 1 to avoid log(0)
```
