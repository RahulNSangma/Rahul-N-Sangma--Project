# Estimation And Confidence Intervals

## A. Build 99% Confidence Interval Using Sample Standard Deviation

```python
import numpy as np
from scipy import stats
```

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)#ddof=1 for sample standard deviation
n = len(data) #sample size
```

```python
sample_mean, sample_std , n
```

```python
confidence_level = 0.99
alpha = 1 - confidence_level
degrees_of_freedom = n - 1
t_critical = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)
```

```python
# As per the Z-Table the value for 0.99 is = 2.33
```

```python
margin_of_error = t_critical * (sample_std / np.sqrt(n))
```

```python
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
```

```python
print("The 99% confidence interval is: ", (confidence_interval))
```

## B. Build 99% Confidence Interval Using Known Population Standard Deviation

```python
import numpy as np
from scipy import stats
```

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
sample_mean
```

```python
sigma = 0.2 # as per given in the question
```

```python
alpha = 0.01  # 1 - confidence level
z_star = stats.norm.ppf(1 - alpha / 2)
```

```python
E = z_star * (sigma / np.sqrt(len(data)))
```

```python
lower_bound = sample_mean - E
upper_bound = sample_mean + E
```

```python
print("The 99% confidence interval is: ", round(lower_bound,2),round(upper_bound,2))
```
