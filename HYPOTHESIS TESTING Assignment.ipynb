# HYPOTHESIS TESTING

```python
# Importing the required modules
import numpy as np
import scipy.stats as stats
from scipy.stats import norm
```

```python
import scipy.stats as stats

# Set the significance level (alpha)
alpha = 0.05

# Calculate the critical value
critical_value = stats.norm.ppf(1 - alpha / 2)  # For a two-tailed test

print("Critical value:", critical_value)
```

```python
z_test_statistic = (950 - 600)/(125/np.sqrt(25))
print(z_test_statistic) 
```

## Conclusion

```python
if((z_test_statistic > critical_value) | (z_test_statistic < (0 - critical_value))):
    print("Reject Ho")
else:
    print("Failed to reject the Ho")
```

# CHI-SQUARE TEST

```python
import scipy.stats as stats
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

```python
crit = stats.chi2.ppf(q = 0.95,df =4)
print(crit)
```

```python
data = np.array([[50, 70],
                    [80, 100],
                    [60, 90],
                    [30, 50],
                    [20, 50]])
```

```python
chi2, p, dof, expected= chi2_contingency(data)
```

```python
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
```

```python
if(chi2 > crit):
    print("Reject Ho")
else:
    print("Failed to reject the Ho")
```
