# 3. Data Visualization:

```python
plt.figure(figsize=(25,10))
sns.boxplot(data=dataset)
plt.title('Box Plot of Cardiotocography Data')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

# Deleding outliers

## Width

```python
Q1 = dataset['Width'].quantile(0.25)
Q3 = dataset['Width'].quantile(0.75)
IQR = Q3 - Q1
min_r = Q1 - (1.5*IQR)
max_r = Q1 + (1.5*IQR)
dataset = dataset[dataset['Width']<=max_r]
```

## MLTV

```python
Q1 = dataset['MLTV'].quantile(0.25)
Q3 = dataset['MLTV'].quantile(0.75)
IQR = Q3 - Q1
min_r = Q1 - (1.5*IQR)
max_r = Q1 + (1.5*IQR)
dataset = dataset[dataset['MLTV']<=max_r]
```

## ALTV

```python
Q1 = dataset['ALTV'].quantile(0.25)
Q3 = dataset['ALTV'].quantile(0.75)
IQR = Q3 - Q1
min_r = Q1 - (1.5*IQR)
max_r = Q1 + (1.5*IQR)
dataset = dataset[dataset['ALTV']<=max_r]
```

## ASTV

```python
Q1 = dataset['ASTV'].quantile(0.25)
Q3 = dataset['ASTV'].quantile(0.75)
IQR = Q3 - Q1
min_r = Q1 - (1.5*IQR)
max_r = Q1 + (1.5*IQR)
dataset = dataset[dataset['ASTV']<=max_r]
```

```python
plt.figure(figsize=(25,10))
sns.boxplot(data=dataset)
plt.title('Box Plot of Cardiotocography Data')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Bar Chart

```python
plt.figure(figsize=(30,10))
sns.countplot(data=dataset)
plt.title('Frequency of Categories for {}'.format(dataset))
plt.xlabel(dataset)
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Pie Chart

```python
categorical_column = 'NSP'

category_counts = dataset[categorical_column].value_counts()
labels = category_counts.index
sizes = category_counts.values

plt.figure(figsize=(30,20))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Frequency of Categories for {}'.format(categorical_column))
plt.axis('equal')
plt.show()
```

# Scatter Plot

```python
variable1 = 'LB'
variable2 = 'AC'

plt.figure(figsize=(8, 6))
sns.scatterplot(x=variable1, y=variable2, data=dataset)
plt.title('Scatter Plot of {} vs. {}'.format(variable1, variable2))
plt.xlabel(variable1)
plt.ylabel(variable2)
plt.show()
```

## Pair Plots

```python
columns_for_pair_plot = ['LB', 'AC', 'FM', 'UC']
sns.pairplot(dataset[columns_for_pair_plot], diag_kind='kde')
plt.show()
```

## Violin Plots

```python
variable_for_violin = 'LB'
grouping_variable = 'NSP'  

plt.figure(figsize=(8, 6))
sns.violinplot(x=grouping_variable, y=variable_for_violin, data=dataset)
plt.title('Violin Plot of {} by {}'.format(variable_for_violin, grouping_variable))
plt.xlabel(grouping_variable)
plt.ylabel(variable_for_violin)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=dataset)
plt.show()
```

# 4. Pattern Recognition and Insights:

```python
# Display of the correlation matrix as a heatmap
correlation_matrix = dataset.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

threshold = 0.7
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            print(f"Strong correlation between {correlation_matrix.columns[i]} and {correlation_matrix.columns[j]}: {correlation_matrix.iloc[i, j]:.2f}")
```
